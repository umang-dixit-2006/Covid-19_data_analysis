# Covid 19 Data Analysis Project

```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("../data+science/data science/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#delete the useless columns
df = dataset.drop(columns=["Lat", "Long"], axis=1, inplace=True)
```

```python
df = dataset.drop(columns=["Province/State"], axis=2, inplace=True)
```

```python
dataset.head()
```

```python
#aggregate the rows by column
corona_dataset_aggregate = dataset.groupby("Country/Region").sum()
corona_dataset_aggregate.head()
```

```python
corona_dataset_aggregate.shape
```

```python
corona_dataset_aggregate.loc["India"].plot()
corona_dataset_aggregate.loc["Spain"].plot()
corona_dataset_aggregate.loc["Italy"].plot()
corona_dataset_aggregate.loc["Russia"].plot()
plt.legend(["India", "Spain", "Italy", "Russia"])
```

```python
#calculate a good measure
corona_dataset_aggregate.loc["India"].plot()
```

```python
corona_dataset_aggregate.loc["India"][:45].plot()
```

```python
#calculate the first derivative of the curve
corona_dataset_aggregate.loc["India"].diff().plot()
```

```python
#calculate the maximum derivative of the curve
corona_dataset_aggregate.loc["India"].diff().max()
```

```python
corona_dataset_aggregate.loc["Russia"].diff().max()
```

```python
corona_dataset_aggregate.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregate.loc["India"].diff().plot()
corona_dataset_aggregate.loc["Russia"].diff().plot()
corona_dataset_aggregate.loc["Italy"].diff().plot()
plt.legend(["India", "Italy", "Russia"])
```

```python
countries = list(corona_dataset_aggregate.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregate.loc[c].diff().max())

corona_dataset_aggregate["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregate
```

```python
#create a new dataframe 
corona_data = pd.DataFrame(corona_dataset_aggregate["Max_infection_rates"])
```
